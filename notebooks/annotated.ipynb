```python
import pandas as pd
```

```python
column_names = [
    'id', #0
    'Offset', #1
    'Text', #2
    'Modality', #8
    'Polarity', #9,
    'Symptom', #10,
    'Date', #11,
    'Entity Relation', #12,
    'Temporal Relation', #13,
    'fk_id', #14
]
```

```python
with open('../data/annotated/journal-2.tsv', encoding='UTF-8') as f:
    document = pd.DataFrame(columns=column_names)
    for i, line in enumerate(f):
        if i in range(7) or line.strip() in ['\n', '']:
            continue
        
        if line.startswith('#'):
            continue
        sentence = line.split('\t')
        
        row = {
            'id': sentence[0] if sentence[0] != '_' else None, #0
            'Offset': sentence[1] if sentence[1] != '_' else None, #1
            'Text': sentence[2] if sentence[2] != '_' else None, #1
            'Modality': sentence[8] if sentence[8] != '_' else None, #8
            'Polarity': sentence[9] if sentence[9] != '_' else None, #9,
            'Symptom': sentence[10] if sentence[10] != '_' else None, #10,
            'Date': sentence[11] if sentence[11] != '_' else None, #11,
            'Entity Relation': sentence[12] if sentence[12] != '_' else None, #12,
            'Temporal Relation': sentence[13] if sentence[13]!= '_' else None, #13,
            'fk_id': sentence[14] if sentence[14] != '_' else None#14
        }
        document.loc[len(document)] = row
```
